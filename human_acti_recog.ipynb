```python
import pandas as pd
```

```python
train_df = pd.read_csv('/content/drive/MyDrive/human activity recognition/train.csv')
test_df = pd.read_csv('/content/drive/MyDrive/human activity recognition/test.csv')
```

```python
train_df.head()
```

```python
x_train = train_df.drop(columns=['Activity'])
y_train = train_df['Activity']
x_test = test_df.drop(columns=['Activity'])
y_test = test_df['Activity']
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score,f1_score,precision_score
```

```python
k = 3
knn = KNeighborsClassifier(n_neighbors=k)
knn.fit(x_train,y_train)
y_pred = knn.predict(x_test)
```

```python
acc =accuracy_score(y_test,y_pred)
print(acc)
f1 = f1_score(y_test,y_pred,average='weighted')
print(f1)
prec = precision_score(y_test,y_pred,average='weighted')
print(prec)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
lorgreg =  LogisticRegression()
lorgreg.fit(x_train,y_train)
y_pred = lorgreg.predict(x_test)


acc =accuracy_score(y_test,y_pred)
print('accuracy:',acc)
f1 = f1_score(y_test,y_pred,average='weighted')
print('f1 score:',f1)
prec = precision_score(y_test,y_pred,average='weighted')
print('precision:',prec)
```
